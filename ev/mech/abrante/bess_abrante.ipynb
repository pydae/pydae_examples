```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import bess
```

```python
m = bess.model()
m.ini({'Ibat':0.0,'K_charger':-1e-6},0.0)
m.report_x()
m.report_y()
#m.report_u()

m.report_params()
m.report_z()
```

```python
m = bess.model()
m.Dt = 0.1
m.decimation = 100
m.ini({'Ibat':0.0,'K_charger':-1e-6, 'SOC0':0.92},0.0)
m.run(    1.0,{})
m.run(20000.0,{'Ibat':10.0})
m.post();

fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(6, 5), dpi=100)

axes[0].plot(m.Time, m.get_values('Vbat_calc'), label=f'Vbat_calc ')
axes[1].plot(m.Time, m.get_values('SOC'), label=f'\nu')

#axes.plot(m.get_values('nu'), m.get_values('snu'), label=f'\nu')

axes[1].set_xlabel('Time (s)')
fig.tight_layout()
```

```python
m = bess.model()
m.Dt = 0.1
m.decimation = 100
times = np.arange(0,30e3,500)

m.ini({'Ibat':0.0,'K_charger':-1e-6, 'SOC0':0.9},0.0)
for t in times:
    m.run(t+300,{'Ibat':30.0})
    m.run(t+500,{'Ibat':-30.0})
m.post();

fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(6, 5), dpi=100)

axes[0].plot(m.Time, m.get_values('Vbat_calc'), label=f'Vbat_calc ')
axes[1].plot(m.Time, m.get_values('SOC'), label=f'\nu')

#axes.plot(m.get_values('nu'), m.get_values('snu'), label=f'\nu')

axes[1].set_xlabel('Time (s)')
fig.tight_layout()
```
