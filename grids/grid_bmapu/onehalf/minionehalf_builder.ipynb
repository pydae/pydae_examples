```python
import numpy as np
import sympy as sym
import pydae.build_cffi as db
```

```python
T_g,H,D,F_n,Droop,H_cis,E_cis_0,K_cis = sym.symbols('T_g,H,D,F_n,Droop,H_cis,E_cis_0,K_cis',real=True) # parameters
p_g,omega,e_cis = sym.symbols('p_g,omega,e_cis',real=True) # dynamic states
p_c,p_l,p_cis = sym.symbols('p_c,p_l,p_cis',real=True)  # 

p_f = -1/Droop*(omega-1)
dp_g = 1/T_g*(p_c + p_f - p_g)
domega = 1/(2*H)*(p_g - p_l + p_cis - D*(omega-1.0))
de_cis = -p_cis/(2*H_cis) + K_cis*(E_cis_0-e_cis)


g_list = []
y_list = []
h_dict = {'p_f':p_f,'p_cis':p_cis}
f_list = [dp_g,domega,de_cis]
x_list = [ p_g, omega, e_cis]

params_dict = {'T_g':1.0,'H':5.0,'D':0.1,'F_n':50.0,'Droop':0.05,'H_cis':2.0, 'E_cis_0':0.5,'K_cis':0.001}
u_dict = {'p_c':0.8, 'p_cis':0.0,'p_l':0.8}

sys = {'name':'minionehalf',
       'params_dict':params_dict,
       'f_list':f_list,
       'g_list':g_list,
       'x_list':x_list,
       'y_ini_list':y_list,
       'y_run_list':y_list,
       'u_run_dict':u_dict,
       'u_ini_dict':u_dict,
       'h_dict':h_dict}

bldr = db.builder(sys)
bldr.build()
```
