```python
import numpy as np
import matplotlib.pyplot as plt
from pydae.plot_tools import set_style
import pydae.ssa as ctrl
colors = set_style(plt)
```

```python
import wf2vsgll 
```

```python
model = wf2vsgll.model()
model.report_u()
model.report_params()
```

```python
model = wf2vsgll.model()

# VSG design
H_v,zeta = 4.0,0.05 # 1.0/np.sqrt(2) 
WB = 2 *np.pi* 50;
R_v,X_v = 0.0,0.3
P_max = 1/X_v
fn = np.sqrt(WB*P_max/(2*H_v))/(2*np.pi)
Dp = 0;
K_i = (2*np.pi*fn)**2/(WB*P_max);
K_g = Dp*K_i;
K_p = (2*zeta*2*np.pi*fn - K_g)/(WB*P_max);
T_q = 1.0/(2*np.pi*10/2)
K_q = (1.0 - 0.0)/(1.05 - 0.95)
K_i_q = 1e-6

params = {
    "K_p_agc":0.0,"K_i_agc":0.0,"K_xif":0.01, # secondary freq. is removed
    "p_ref_G01":0.8,"q_ref_G01":0.0,
    "p_ref_G02":0.8,"q_ref_G02":0.0,
    'R_v_G01':R_v,'X_v_G01':X_v,'K_p_G01':K_p,'K_i_G01':K_i,'K_g_G01':K_g,'K_q_G01':K_q,'T_q_G01':T_q,'K_i_q_G01':K_i_q,
    'R_v_G02':R_v,'X_v_G02':X_v,'K_p_G02':K_p,'K_i_G02':K_i,'K_g_G02':K_g,'K_q_G02':K_q,'T_q_G02':T_q,'K_i_q_G02':K_i_q
}
model.itol = 1e-6
model.max_it = 1000

model.ini(params,'xy_0_2.json')
#model.ini(params,'xy_0_2.json')
#model.save_xy_0('xy_0.json')
model.report_x()
model.report_y()
model.report_z()
```

```python
ctrl.A_eval(model)
ctrl.damp_report(model)
```

```python
model.ini(params,'xy_0.json')
model.run(1.0,{})
model.run(2.0,{'p_ref_G01':0.7})
model.run(10.0,{})
model.post();
```

```python
plt.close('all')
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(8, 3), frameon=False, dpi=100, squeeze=False)

axes[0,0].plot(model.Time, model.get_values('p_G01'), label="$p_{G01}$")
axes[0,0].plot(model.Time, model.get_values('p_G02'), label="$p_{G02}$")

axes[0,1].plot(model.Time, 50*(model.get_values('omega_v_G01')-1), label="$\omega_{G01}$")
axes[0,1].plot(model.Time, 50*(model.get_values('omega_v_G02')-1), label="$\omega_{G02}$")

for ax in axes.flatten(): 
    ax.legend()
```
