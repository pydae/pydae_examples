```python
model = sofc.model()
model.ini({'i_dc_ref':150, 'U_opt':0.7},1)
model.report_x()
model.report_y()
model.report_z()

for item in model.x_list:
    print(f'{item} = {model.get_value(item)}')
```

```python
model = sofc.model()
model.ini({'i_dc_ref':150, 'U_opt':0.7},1)
model.Dt = 0.1
model.decimation = 1
model.run( 100.0,{})
model.run(1000.0,{'i_dc_ref':200})
model.post();

fig,axes = plt.subplots(nrows=3)

axes[0].plot(model.Time,  model.get_values('i_dc'),label='i_dc')
axes[1].plot(model.Time,  model.get_values('v_dc'),label='v_dc')

axes[2].plot(model.Time,  model.get_values('p_h2'),label='p_h2')
axes[2].plot(model.Time,  model.get_values('p_h2o'),label='p_h2o')
axes[2].plot(model.Time,  model.get_values('p_o2'),label='p_o2')

for ax in axes:
    ax.legend()
    ax.grid()
    #ax.set_xlim([0,20])
ax.set_xlabel('Time (s)')
fig.tight_layout()
```
